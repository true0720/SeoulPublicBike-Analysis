```python
# ğŸ“¦ ë”°ë¦‰ì´ ë°ì´í„° ë¶„ì„ Colab í…œí”Œë¦¿ (ëª¨ë“  ì œê³µ íŒŒì¼ ë°˜ì˜)

# 1. ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, shapiro

# 2. íŒŒì¼ ì—…ë¡œë“œ (ì—¬ëŸ¬ íŒŒì¼ ì§€ì›)
from google.colab import files
uploaded = files.upload()

# 3. íŒŒì¼ë³„ ì½ê¸°
main_df, summary_df, station_df, region_df = None, None, None, None
for fname in uploaded:
    if 'tpss_bcycl' in fname:
        for enc in ['cp949', 'utf-8', 'euc-kr']:
            try:
                main_df = pd.read_csv(fname, encoding=enc)
                break
            except:
                continue
    elif 'rent_summary' in fname:
        summary_df = pd.read_csv(fname, encoding='cp949')
    elif 'stations' in fname:
        station_df = pd.read_excel(fname)
    elif 'ìì¹˜êµ¬' in fname:
        region_df = pd.read_excel(fname)

# 4. ì»¬ëŸ¼ ì •ë¦¬ (ë©”ì¸ ë¶„ì„ ë°ì´í„°)
df = main_df.copy()
df.rename(columns={
    'ê¸°ì¤€_ë‚ ì§œ': 'RENT_DATE',
    'ê¸°ì¤€_ì‹œê°„ëŒ€': 'RENT_TIME',
    'ì‹œì‘_ëŒ€ì—¬ì†Œëª…': 'RENT_STATION_NAME',
    'ì¢…ë£Œ_ëŒ€ì—¬ì†Œëª…': 'RETURN_STATION_NAME',
    'ì „ì²´_ê±´ìˆ˜': 'RENT_COUNT',
    'ì „ì²´_ì´ìš©_ë¶„': 'RENT_DURATION_MIN',
    'ì „ì²´_ì´ìš©_ê±°ë¦¬': 'RENT_DISTANCE_METER'
}, inplace=True)

# ë‚ ì§œ/ì‹œê°„ ì²˜ë¦¬
df['RENT_DATE'] = pd.to_datetime(df['RENT_DATE'], format='%Y%m%d')
df['hour'] = df['RENT_TIME'].astype(str).str.zfill(4).str[:2].astype(int)
df['weekday'] = df['RENT_DATE'].dt.dayofweek

# íŒŒìƒ ë³€ìˆ˜ ìƒì„±
day_name = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df['day_name'] = df['weekday'].apply(lambda x: day_name[x])
df['month'] = df['RENT_DATE'].dt.month
df['season'] = df['month'].map({1:'Winter',2:'Winter',3:'Spring',4:'Spring',5:'Spring',6:'Summer',7:'Summer',8:'Summer',9:'Fall',10:'Fall',11:'Fall',12:'Winter'})
df['is_weekend'] = df['weekday'] >= 5

# ì´ìƒì¹˜ ì œê±°
df.dropna(subset=['RENT_DURATION_MIN', 'RENT_DISTANCE_METER'], inplace=True)
df = df[(df['RENT_DURATION_MIN'] > 0) & (df['RENT_DISTANCE_METER'] > 0)]

# ë¶„ì„ 1: ê¸°ì´ˆ í†µê³„
print("\n[ê¸°ì´ˆ í†µê³„ ìš”ì•½]")
print(df[['RENT_COUNT', 'RENT_DURATION_MIN', 'RENT_DISTANCE_METER']].describe())

# ë¶„ì„ 2: ì‹œê°„ëŒ€/ìš”ì¼/ì›”ë³„ ì‹œê°í™”
df['time_slot'] = pd.cut(df['hour'], bins=[0,6,9,12,15,18,21,24], right=False,
                         labels=['00~06','06~09','09~12','12~15','15~18','18~21','21~24'])
df['time_slot'].value_counts().sort_index().plot(kind='bar', title='ì‹œê°„ëŒ€ë³„ ëŒ€ì—¬/ë°˜ë‚© ë¶„í¬')
plt.xlabel('ì‹œê°„ëŒ€'); plt.ylabel('ê±´ìˆ˜'); plt.show()

df['day_name'].value_counts().reindex(day_name).plot(kind='bar', title='ìš”ì¼ë³„ ëŒ€ì—¬ëŸ‰')
plt.show()
df['month'].value_counts().sort_index().plot(kind='bar', title='ì›”ë³„ ëŒ€ì—¬ëŸ‰')
plt.show()

# ë¶„ì„ 3: ë¶„í¬ + ì •ê·œì„± ê²€ì •
print("\n[ì´ìš©ì‹œê°„ ì •ê·œì„± ê²€ì •]")
print("ì™œë„:", skew(df['RENT_DURATION_MIN']))
print("ì²¨ë„:", kurtosis(df['RENT_DURATION_MIN']))
print("Shapiro-Wilk p-value:", shapiro(df['RENT_DURATION_MIN'].sample(500, random_state=42)).pvalue)
sns.histplot(df['RENT_DURATION_MIN'], bins=100, kde=True)
plt.title('ì´ìš© ì‹œê°„ ë¶„í¬')
plt.show()

# ë¶„ì„ 4: ì´ìƒì¹˜ ì‹œê°í™”
sns.boxplot(x=df['RENT_DURATION_MIN'])
plt.title('ì´ìš© ì‹œê°„ ì´ìƒì¹˜')
plt.show()

# ë¶„ì„ 5: ì£¼ë§/ê³„ì ˆ ë¶„ì„
sns.barplot(x='is_weekend', y='RENT_COUNT', data=df)
plt.xticks([0,1], ['í‰ì¼', 'ì£¼ë§'])
plt.title('ì£¼ë§ vs í‰ì¼ ëŒ€ì—¬ëŸ‰')
plt.show()
sns.boxplot(x='season', y='RENT_DURATION_MIN', data=df)
plt.title('ê³„ì ˆë³„ ì´ìš© ì‹œê°„')
plt.show()

# ë¶„ì„ 6: ì¸ê¸° ëŒ€ì—¬ì†Œ
popular = df.groupby('RENT_STATION_NAME')['RENT_COUNT'].sum().sort_values(ascending=False).head(10)
popular.plot(kind='barh', title='ì¸ê¸° ëŒ€ì—¬ì†Œ TOP 10', color='green')
plt.gca().invert_yaxis()
plt.show()

# ë¶„ì„ 7: ë³´ì¡° ë°ì´í„° í™•ì¸
if region_df is not None:
    print("\n[ìì¹˜êµ¬ë³„ ëŒ€ì—¬ê±´ìˆ˜ ìš”ì•½]")
    print(region_df.head())

if summary_df is not None:
    print("\n[ëŒ€ì—¬ì†Œ ì‹œê°„ëŒ€ ìš”ì•½] íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸° ì™„ë£Œ")
    print(summary_df.head())

if station_df is not None:
    print("\n[ì¸ê¸° ëŒ€ì—¬ì†Œ ì •ë ¬ ê²°ê³¼] íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸° ì™„ë£Œ")
    print(station_df.head())

# ë!
```
